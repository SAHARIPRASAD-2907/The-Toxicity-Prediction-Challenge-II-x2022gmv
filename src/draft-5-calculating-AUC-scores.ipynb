```python
import pandas as pd
import numpy as np 
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
```

```python
train_data = pd.read_csv("./smile_description_train.csv")
test_data = pd.read_csv("./smile_description_test.csv")
```

```python
len(train_data)
```

```python
train_data =  train_data.dropna()
```

```python
train_data
```

```python
AUC_SCORES = {
}
```

```python
for col in train_data:
    print(col)
    X = pd.DataFrame(train_data[col])
    y = train_data["label"]
    print(y.shape)
    model =  RandomForestClassifier()
    model.fit(X,y)
    prob = model.predict_proba(X)[:, 1]
    auc = roc_auc_score(y,prob)
    if auc<0.5:
        auc = 1 - auc
    AUC_SCORES[col] = auc
    
     
```

```python
y = train_data["label"]

y
```

```python
auc_scores = AUC_SCORES.values()
columns = AUC_SCORES.keys()

auc_data_frame = pd.DataFrame()
auc_data_frame["Columns"] = columns
auc_data_frame["AUC_VALUES"] = auc_scores
```

```python
auc_data_frame = auc_data_frame.sort_values(by=["AUC_VALUES"],ascending=False)
```

```python
auc_data_frame.to_csv("AUC_VALUES.csv")
```
