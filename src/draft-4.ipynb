```python
import pandas as pd
import numpy as np 
```

```python
train_data = pd.read_csv("./detail_smile_data_train.csv")
test_data = pd.read_csv("./detail_smile_data_test.csv")
```

```python
X = train_data.drop(["Assay_Id","label"],axis=1)
y = train_data["label"]
```

```python
print(len(y))
```

```python
print(len(X))
```

```python
model.fit(X,y)
```
