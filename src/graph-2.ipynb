```python
import pandas as pd
import numpy as np 
```

```python
graph_data = pd.read_csv("./scores.csv")
```

```python
y = graph_data["Index"]
x1 = graph_data["F1 score-local"]*100
x2 = graph_data["F1 score-global"]*100
```

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
plt.plot(y, x2, label = "Kaagle score",marker="x")
plt.plot(y, x1, label = "Local score",marker="o")
plt.axvline(55,color="g",label="Submission after presentation 1")
plt.axvline(94,color="y",label="Best submission")
plt.legend()
plt.xlabel("Submission number")
plt.ylabel("F1 score")
plt.title("Presentation 2 (x2022gmv)")
plt.show()
```
